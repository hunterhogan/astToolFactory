# load dataframe

```python
from astToolFactory import settingsManufacturing
import pandas

dataframe: pandas.DataFrame = pandas.read_pickle(settingsManufacturing.pathFilenameDataframeAST)
dataframe = dataframe.drop(columns=['list4TypeAlias_subcategoryNames'])
```

```python
from astToolkit import identifierDotAttribute
from hunterMakesPy import raiseIfNone
from itertools import chain
from pprint import pprint
from typing import Any
import ast

dictionary_astClasses: dict[str, Any] = {}

for astClass in [C for C in [ast.AST, *chain(*(c.__subclasses__() for c in [ast.AST, ast.Constant, *ast.AST.__subclasses__()]))] if issubclass(C, ast.AST)]:
	identifierClass: str = astClass.__name__
	identifierDotClass: identifierDotAttribute = f'`ast.{identifierClass}`'
	subclassList: list[str] = sorted([c.__name__ for c in astClass.__subclasses__() if issubclass(c, ast.AST)], key=lambda s: s.lower())
	matchesClasses: list[identifierDotAttribute] = list(dict.fromkeys([identifierDotClass, *[f'`ast.{name}`' for name in subclassList]]))
	parentClass: identifierDotAttribute = raiseIfNone(raiseIfNone(astClass.__base__).__name__)

	dictionary_astClasses[identifierClass] = {
		'ClassDefIdentifier': identifierClass,
		'versionMajorPythonInterpreter': 3,
		'versionMinorPythonInterpreter': 14,
		'versionMicroPythonInterpreter': 2,
		'base': parentClass,
	}

del dictionary_astClasses['EnhancedAST']

list_astClasses = list(dictionary_astClasses.values())
# pprint(list_astClasses, width=200, compact=True)
```

```python
# add
dataframe = pandas.concat([dataframe, pandas.DataFrame(list_astClasses)], ignore_index=True)
```

```python
# Fill missing values in 'kwarg_annotationIdentifier' column with 'No'
dataframe['kwarg_annotationIdentifier'] = dataframe['kwarg_annotationIdentifier'].fillna('No')
```

# save dataframe

```python
dataframe.to_pickle(settingsManufacturing.pathFilenameDataframeAST)
```
